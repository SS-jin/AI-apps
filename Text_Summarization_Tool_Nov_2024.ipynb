```python
import os
import torch
from transformers import BartTokenizer, BartForConditionalGeneration
from google.colab import files
from PyPDF2 import PdfReader

# Function to clean and preprocess extracted text
def clean_text(text):
    # Remove unnecessary whitespace and newlines
    text = ' '.join(text.split())
    return text

# Function to split text into paragraphs
def split_into_paragraphs(text):
    paragraphs = text.split('\n')
    paragraphs = [p.strip() for p in paragraphs if p.strip()]
    return paragraphs

# Function to summarize text with dynamic max_new_tokens and min_new_tokens based on percentages
def summarize(text, max_length_percent, min_length_percent):
    # Clean the text
    text = clean_text(text)

    # Calculate max_new_tokens and min_new_tokens based on the original text length
    text_length = len(text.split())
    max_new_tokens = min(1024, int(text_length * max_length_percent / 100))
    min_new_tokens = int(text_length * min_length_percent / 100)

    inputs = tokenizer.encode(text, return_tensors="pt", max_length=1024, truncation=True)
    outputs = model.generate(
        inputs,
        max_new_tokens=max_new_tokens,  # Dynamic max_new_tokens based on percentage
        min_length=min_new_tokens,  # Dynamic min_length based on percentage
        length_penalty=2.0,
        num_beams=6,  # Increased number of beams for more thorough search
        early_stopping=True
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Function to read text from a text file
def read_text_file(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            return file.read()
    except UnicodeDecodeError:
        try:
            with open(filename, 'r', encoding='latin-1') as file:
                return file.read()
        except UnicodeDecodeError:
            with open(filename, 'r', encoding='utf-16') as file:
                return file.read()

# Function to read text from a PDF file
def read_pdf_file(filename):
    text = ""
    try:
        with open(filename, 'rb') as file:
            reader = PdfReader(file)
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text += page.extract_text()
    except Exception as e:
        print(f"Error reading PDF file: {e}")
    return text

# Load BART model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

# Check if a GPU is available and use it if possible
device = 0 if torch.cuda.is_available() else -1

# Upload file
uploaded = files.upload()

# Read the uploaded file (either .txt or .pdf)
filename = list(uploaded.keys())[0]
if filename.endswith('.txt'):
    text = read_text_file(filename)
elif filename.endswith('.pdf'):
    text = read_pdf_file(filename)
else:
    text = ""
    print("Unsupported file type. Please upload a .txt or .pdf file.")

# Set the max_length and min_length as percentages
max_length_percent = 30  # Example: set max length to 30% of the original text length
min_length_percent = 20   # Example: set min length to 20% of the original text length

# Generate and print the summary if text was extracted
if text:
    # Split text into paragraphs
    paragraphs = split_into_paragraphs(text)

    # Summarize each paragraph or group of paragraphs
    chunk_size = 1024  # Define the chunk size limit for the model
    chunks = []
    current_chunk = ""

    for paragraph in paragraphs:
        if len(current_chunk) + len(paragraph) < chunk_size:
            current_chunk += " " + paragraph
        else:
            chunks.append(current_chunk)
            current_chunk = paragraph

    if current_chunk:
        chunks.append(current_chunk)

    # Summarize each chunk
    summaries = [summarize(chunk, max_length_percent, min_length_percent) for chunk in chunks]
    combined_summary = " ".join(summaries)

    # Further summarize the combined summary if it's still too long
    final_summary = summarize(combined_summary, max_length_percent, min_length_percent)
    print("Summary: ", final_summary)
else:
    print("No text found to summarize.")
```
